```python
import numpy as np
test = np.array([[0,1,2],[1,0,0],[2,0,1]])
```

```python
_, width = test.shape
indices_y, indices_x = np.indices(test.shape)
indices_shifted = indices_x + test
is_clipped = indices_shifted >= width
indices_shifted[is_clipped] = width - 1
not_clipped = np.logical_not(is_clipped)
test[not_clipped], test[indices_y, indices_shifted][not_clipped]
```
